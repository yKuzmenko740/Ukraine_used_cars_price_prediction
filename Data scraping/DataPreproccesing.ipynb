```python
def hasukr(s):
    lower = set('абвгґдеєжзиіїйклмнопрстуфхцчшщьюя')
    return lower.intersection(s.lower()) != set()
```

```python
def dropping(data:pd.DataFrame, *args , **kwargs):
    print(f"Shape before dropping: {data.shape}")
    data = data.drop(*args, **kwargs)
    print(f"Shape after dropping: {data.shape}")
    return data
```

```python
ukr_indexes = df[df.Brand.apply(lambda x: hasukr(x))].index
df = dropping(df, ukr_indexes)
```

```python
cars = df.Brand.value_counts()[df.Brand.value_counts() < 200].index
small_cars = df[df.Brand.isin(cars)].index
small_cars
```

```python
df = dropping(df, small_cars)
```

```python
def transform_benz(data):
    dct = {"Дизель": 'dis', 'Бензин': 'benz', 'Газ / Бензин': 'gb', "Газ": 'gas', 'Гібрид': 'gibr', 'Електро': 'elctr', 'Газ пропан-бутан':'gpb', 'Газ метан': 'gb'}
    return data.replace({'Benz': dct})
```

```python
def transform_trans(data):
    dct = {'Автомат':'auto', 'Ручна/Механіка': 'mech', 'Варіатор':'var','Типтронік': 'tip', 'Робот': 'rob'}
    return data.replace({'Transmission': dct})
```
